```python
# Initialize Antropic Chat client
from anthropic import Anthropic
client = Anthropic()
```

```python
# Helper Chat Functions


## Function to Append User messasge
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


## Function to Append Assistant messasge
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


## Main Function for chat
def get_stream_response(messages, system=None, temperature=0.1):

    model = "claude-sonnet-4-0"  # you can change the model according to your usecase
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
    }
    if system:
        params["system"] = system
    with client.messages.stream(
        model=model, max_tokens=100, messages=messages
    ) as stream:
        for text in stream.text_stream:
            print(text)
            pass
    return stream.get_final_message()


def get_chat_content(message):
    return message.content[0].text
```

```python
# Global list of messages
messages =[]
```

```python
def chat_stream(message, system=None, temperature=0.1):
    # Add in the initial user question of "Define quantum computing in one sentence"
    add_user_message(messages=messages, text=message)

    # Pass the list of messages into 'chat' to get an answer
    chatResponse = get_stream_response(
        messages=messages,
        system=system,
        temperature=temperature,
    )

    responseText = get_chat_content(chatResponse)
    # Take the answer and add it as an assistant message into our list
    add_assistant_message(messages=messages, text=responseText)
```
