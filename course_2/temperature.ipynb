```python
# Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Initialize Antropic Chat client
from anthropic import Anthropic
client = Anthropic()
```

```python
# Helper Chat Functions


## Function to Append User messasge
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


## Function to Append Assistant messasge
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


## Main Function for chat
def get_response(messages, system=None, temperature=0.0):

    model = "claude-sonnet-4-0"  # you can change the model according to your usecase
    params = {
        "model": model,
        "max_tokens": 100,
        "messages": messages,
        "temperature": temperature,
    }
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message


def get_chat_content(message):
    return message.content[0].text
```

```python
def checkTempResponse(temp=0.1):
    messages=[]
    add_user_message(messages,"suggest best indian movies")
    response = get_response(messages=messages,temperature=temp)
    return get_chat_content(response)
```

```python
# With less temprature we will likely get the almost same responses each time 
checkTempResponse(temp=0.1)
```

```python

# With more temprature we will get different responses each time 
# Max temperature value = 1.0
checkTempResponse(temp=1.0)
```
