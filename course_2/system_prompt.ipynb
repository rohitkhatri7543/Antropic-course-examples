```python
# Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Initialize Antropic Chat client
from anthropic import Anthropic
client = Anthropic()
```

```python
# Helper Chat Functions


## Function to Append User messasge
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


## Function to Append Assistant messasge
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


## Main Function for chat
def get_response(messages, system=None):

    model = "claude-sonnet-4-0"  # you can change the model according to your usecase
    params = {"model": model, "max_tokens": 100, "messages": messages}
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message


def get_chat_content(message):
    return message.content[0].text
```

```python
# Global list of messages
messages =[]

def chat(message, system=None):
    # Add in the initial user question of "Define quantum computing in one sentence"
    add_user_message(messages=messages,text=message)

    # Pass the list of messages into 'chat' to get an answer
    chatResponse = get_response(messages=messages,system=system)  

    responseText = get_chat_content(chatResponse)
    # Take the answer and add it as an assistant message into our list
    add_assistant_message(messages=messages, text=responseText)

    return responseText
    
```

```python
# Interactive AI chat
# system = """
# You are a patient math tutor.
# Do not directly answer a student's questions.
# Guide them to a solution step by step.
# """


while True:
    user_input = input("type your message or click 'esc' button to exit..")
    if user_input == None or user_input == "":
        print("\nUser Exited!")
        break
    response = chat(user_input)
    print(f"\nUser: {user_input}\n")
    print(f"AI Response: {response}")
```
